# Polynomial Regression and Pipelines

```python
#import
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression

from sklearn.pipeline import Pipeline
```

```python
#read_data
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(path)
df.head()
```

```python
#PolynomialFeatures
pr=PolynomialFeatures(degree=2 , include_bias=False)
#fit
x_polly=pr.fit_transform(df[['horsepower','curb-weight']])
x_polly
```

```python
# normalixe 
SCALE =StandardScaler()
SCALE.fit(df[['horsepower','highway-mpg']])
```

```python
x_scale=SCALE.transform(df[['horsepower','highway-mpg']])
x_scale
```

### Piplines Constructor

```python
Input=[('scala',StandardScaler()),('polynomial',PolynomialFeatures(degree=2)),('mode',LinearRegression())]
```

```python
#Pipline consructor
pipi=Pipeline(Input)
pipi
```

```python
#train pipline object
pipi.fit(df[['horsepower','curb-weight','engine-size','highway-mpg']],df.price)
```
