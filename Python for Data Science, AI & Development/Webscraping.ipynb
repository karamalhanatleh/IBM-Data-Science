## Overview Python Program

```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Muhammad").text
```

```python
soup = BeautifulSoup(page,'html.parser')
```

```python
soup
```

```python
artists = soup.find_all("a")
artists
```

```python
for art in artists:
    name=art.contents[0]
    fullink=art.get('href')
    print(name)
    print(fullink)
    print('_'*10)
```
