# Basic Plotting with Matplotlib


```python
#import packages
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(5,5,'o')
plt.show()
```

```python
plt.plot(5,5,'o')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("plotting Examples")
plt.show()
```

```python
#creat dataframe india_china_df
india_china_df = pd.DataFrame({
    "indea":[8880 , 8670 , 8147 , 7338 , 5704],
    "China":[5123 , 6682 , 3308 , 1863 , 1527],
    "index":[1980 , 1981 , 1982 , 1983  , 1984]
    
})

india_china_df = india_china_df.set_index('index')
india_china_df
```

```python
#plot line
india_china_df.plot(kind="line")
```

```python
#plot hist
india_china_df.plot(kind="hist")
```

```python
#plot hist India
india_china_df['indea'].plot(kind="hist")
```
