# Histograms

```python
#import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(10,30,20).tolist()
y=np.random.randint(50,60,20).tolist()
z=np.random.randint(80,90,20).tolist()
s=x+y+z
df=pd.DataFrame()
df['age']=s
df.age.hist()
```

```python
df_can=pd.read_csv('canda.csv',index_col=['Country'])
df_can
```

```python
df_can['2013'].plot(kind='hist')
```

```python
count , bin_edges = np.histogram(df_can['2013'])
df_can['2013'].plot(kind='hist' ,xticks=bin_edges)


plt.title('Histogram of Immigration from 195 countries in 2013') # add a title to the histogram
plt.ylabel('Number of Countries') # add y-label
plt.xlabel('Number of Immigrants') # add x-label

plt.show()
```
