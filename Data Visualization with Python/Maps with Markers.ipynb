# Maps with Markers


```python
import folium
```

```python
canada_map=folium.Map(
location=[56.130 , -106.35],
zoom_start=4)

ontario =folium.map.FeatureGroup()
canada_map
```

```python
canada_map=folium.Map(
location=[56.130 , -106.35],
zoom_start=4)

ontario =folium.map.FeatureGroup()

ontario.add_child(
                    folium.features.CircleMarker(
                        [51.25 , -85.32] , radius=5,
                            color="red" , fill_color="red"
                    )
                    )
canada_map.add_child(ontario)

folium.Marker([51.25 , -85.32] , popup="ontario").add_to(canada_map)

canada_map
```

```python
canada_map=folium.Map(
location=[32.193402 , 35.8370178],
zoom_start=30)

ontario=folium.map.FeatureGroup()

ontario.add_child(folium.features.CircleMarker(
                        [32.193402 , 35.8370178],
                            radius=10, color="red"
))
canada_map.add_child(ontario)
folium.Marker([32.193402 , 35.8370178]).add_to(canada_map)
canada_map
```
